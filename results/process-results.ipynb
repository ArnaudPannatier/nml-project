```python
import pandas as pd
df = pd.read_csv("results.csv")
df["model"] = df["model"].apply(lambda s: s[:-1])
df["date"] = pd.to_datetime(df["time"]).dt.day
```

```python
df = df[df["date"]==6]
```

```python
min_df = df.groupby(["model", "seed"]).min()
min_df.groupby("model").mean()
```

```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
min_df.groupby("model").std()
```
