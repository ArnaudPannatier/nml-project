```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
plt.gray()
import numpy as np
import torch
from positional_encoding import SinCosPositionalEncoding

pe = SinCosPositionalEncoding(20)
x = torch.arange(0,200)
pe(x).shape
```

```python
plt.imshow(pe(x).T.numpy())
```

```python
X,Y = torch.meshgrid(x,x,indexing="ij")
```

```python
fig, axs= plt.subplots(3,3)

pe_x, pe_y = pe(X), pe(Y)
print(pe_x.shape)

for ax, idx in zip(axs,[0,11,-1]):
    ax[0].imshow(pe_x[...,idx])
    ax[1].imshow(pe_y[...,idx])
    ax[2].imshow(pe_x[...,idx]+pe_y[...,idx])

```
